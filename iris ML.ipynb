```python
## Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
## Import and explore data
dataset = pd.read_csv("Iris.csv")
dataset.describe()
dataset.shape

## Check null values
dataset.isnull()
dataset.drop('Id', axis=1, inplace=True)

dataset.head()
```

```python
sns.countplot(x='Species',data=dataset)
plt.title('Species',fontsize=20)
plt.show()
```

```python
dataset.plot(kind="scatter",x='SepalLengthCm',y='PetalLengthCm')
plt.show()
```

```python
sns.scatterplot(x = dataset["SepalLengthCm"],y = dataset["SepalWidthCm"]
                ,hue = dataset["Species"])
plt.show()
```

```python
sns.scatterplot(x = dataset["PetalLengthCm"],y = dataset["PetalWidthCm"]
                ,hue = dataset["Species"])
plt.show()
```

```python
## K-Means clustering
X = dataset.iloc[:, [0,1,2,3]].values
print(X)

wcss=[]

for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.title(' Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans= KMeans(n_clusters = 3 , init = 'k-means++', random_state = 42)
predict_kmeans = kmeans.fit_predict(X)
print(predict_kmeans)
```

```python
plt.scatter(X[predict_kmeans == 0, 0], X[predict_kmeans == 0, 1], 
            s = 50, c = 'red', label = 'Iris-setosa')
plt.scatter(X[predict_kmeans == 1, 0], X[predict_kmeans == 1, 1], 
            s = 50, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[predict_kmeans == 2, 0], X[predict_kmeans == 2, 1],
            s = 50, c = 'green', label = 'Iris-virginica')
```

```python
# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 75, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Iris Species')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
plt.show()
```
